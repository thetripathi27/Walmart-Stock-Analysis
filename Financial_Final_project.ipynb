```python
def implement_bb_strategy(data, lower_bb, upper_bb):
    buy_price = []
    sell_price = []
    bb_signal = []
    signal = 0

    for i in range(len(data)):
        if data[i-1] > lower_bb[i-1] and data[i] < lower_bb[i]:
            if signal != 1:
                buy_price.append(data[i])
                sell_price.append(np.nan)
                signal = 1
                bb_signal.append(signal)
            else:
                buy_price.append(np.nan)
                sell_price.append(np.nan)
                bb_signal.append(0)
        elif data[i-1] < upper_bb[i-1] and data[i] > upper_bb[i]:
            if signal != -1:
                buy_price.append(np.nan)
                sell_price.append(data[i])
                signal = -1
                bb_signal.append(signal)
            else:
                buy_price.append(np.nan)
                sell_price.append(np.nan)
                bb_signal.append(0)
        else:
            buy_price.append(np.nan)
            sell_price.append(np.nan)
            bb_signal.append(0)

    return buy_price, sell_price, bb_signal

buy_price, sell_price, bb_signal = implement_bb_strategy(df['Close'], df['lower_bb'], df['upper_bb'])
```

```python
def calculate_wacc(market_value_of_equity, market_value_of_debt, cost_of_equity, cost_of_debt, corporate_tax_rate):
    """
    Calculate the Weighted Average Cost of Capital (WACC).

    Parameters:
    - market_value_of_equity (float): Market value of the company's equity (E).
    - market_value_of_debt (float): Market value of the company's debt (D).
    - cost_of_equity (float): Cost of equity (Re).
    - cost_of_debt (float): Cost of debt (Rd).
    - corporate_tax_rate (float): Corporate tax rate (Tc).

    Returns:
    - float: The WACC of the company.
    """
    total_value = market_value_of_equity + market_value_of_debt
    equity_ratio = market_value_of_equity / total_value
    debt_ratio = market_value_of_debt / total_value

    wacc = (equity_ratio * cost_of_equity) + (debt_ratio * cost_of_debt * (1 - corporate_tax_rate))
    return wacc


wacc_value = calculate_wacc(market_value_of_equity = equity,
                            market_value_of_debt = debt,
                            cost_of_equity = cost_of_equity,
                            cost_of_debt = cost_of_debt,
                            corporate_tax_rate = income_tax_rate)
print(f"The WACC for Walmart is: {wacc_value:.4f}")
```

# Monte Carlo Simulation

```python
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
import dash
from dash import dcc, html, Input, Output
import plotly.graph_objs as go

# Set up Dash app
app = dash.Dash(__name__)

# Define function to fetch stock data and perform Monte Carlo simulation
def perform_simulation(ticker, n_simulations, n_days):
    # Downloading the last year's stock data for the specified ticker
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)
    data = yf.download(ticker, start=start_date, end=end_date)

    # Calculating daily returns
    data['Returns'] = data['Adj Close'].pct_change()

    last_price = data['Adj Close'][-1]

    simulation_df = pd.DataFrame()

    for _ in range(n_simulations):
        daily_volatility = data['Returns'].std()

        price_series = [last_price]

        for _ in range(n_days):
            price = price_series[-1] * (1 + np.random.normal(0, daily_volatility))
            price_series.append(price)

        simulation_df[_] = price_series

    return simulation_df, last_price

# Define layout
app.layout = html.Div([
    html.H1("Monte Carlo Simulation for Stock Price"),
    dcc.Input(id='ticker-input', type='text', placeholder='Enter Ticker Symbol', value='WMT'),
    html.Button('Run Simulation', id='run-button', n_clicks=0),
    dcc.Graph(id='simulation-plot'),
    html.Div(id='simulation-parameters')
])

# Define callback to update the plot and display simulation parameters based on user input
@app.callback(
    [Output('simulation-plot', 'figure'),
     Output('simulation-parameters', 'children')],
    [Input('run-button', 'n_clicks')],
    [Input('ticker-input', 'value')]
)
def update_simulation_plot(n_clicks, ticker):
    if n_clicks is None or n_clicks == 0:
        # Return blank plot and empty div if no simulation has been run yet
        return {}, []

    # Run simulation if the button has been clicked
    simulation_df, last_price = perform_simulation(ticker, n_simulations=100, n_days=30)

    traces = [go.Scatter(x=simulation_df.index, y=simulation_df[col], mode='lines', name=f'Simulation {col+1}') for col in simulation_df.columns]
    traces.append(go.Scatter(x=simulation_df.index, y=[last_price] * len(simulation_df), mode='lines', name='Last Price', line=dict(color='red', dash='dash')))

    layout = {
        'title': f'Monte Carlo Simulation for {ticker} Stock',
        'xaxis': {'title': 'Day'},
        'yaxis': {'title': 'Price'}
    }

    # Display simulation parameters
    parameters_text = [
        html.P(f'Ticker Symbol: {ticker}'),
        html.P('Number of Simulations: 100'),
        html.P('Number of Days: 30')
    ]

    return {'data': traces, 'layout': layout}, parameters_text

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime, timedelta

# Downloading the last year's stock data
end_date = datetime.now()
start_date = end_date - timedelta(days=365)

# Fetching data
data = yf.download(ticker, start=start_date, end=end_date)

data['Adj Close'].plot(figsize = (10,6), title = f"Last year Stock Price chart of {ticker}")

# Calculating daily returns
data['Returns'] = data['Adj Close'].pct_change()

# Setting the number of simulations and the number of days to project
n_simulations = 100
n_days = 30

last_price = data['Adj Close'][-1]

simulation_df = pd.DataFrame()

for x in range(n_simulations):
    count = 0
    daily_volatility = data['Returns'].std()

    price_series = []

    # Generating the price list for the next year
    price = last_price * (1 + np.random.normal(0, daily_volatility))
    price_series.append(price)

    for y in range(n_days):
        if count == n_days+1:
            break
        price = price_series[count] * (1 + np.random.normal(0, daily_volatility))
        price_series.append(price)
        count += 1

    simulation_df[x] = price_series

# Plotting the simulation
plt.figure(figsize=(10,6))
plt.plot(simulation_df)
plt.axhline(y=last_price, color='r', linestyle='-')
plt.title(f'Monte Carlo Simulation for {ticker} Stock')
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```
